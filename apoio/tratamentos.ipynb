```python
import pandas as pd
import datetime
from unidecode import unidecode
```

```python
format_columns = lambda column: unidecode(column.lower())
```

```python
parse_date = lambda dates: datetime.datetime.strptime('%m/%d/%Y', dates)
```

```python
df = pd.read_csv('daily-treasure.csv')
```

```python
df.columns = df.columns.map(format_columns)
```

```python
def format_date(date):
  splitted_date = date.split('/')

  return f"{splitted_date[2]}-{splitted_date[0]}-{splitted_date[1]}"
```

```python
df['date'] = df['date'].apply(format_date)
```

```python
df.columns = df.columns.map(lambda col: col.replace(' ', '_'))
```

```python
df.to_excel('daily_treasure.xlsx')
```

```python
dfCPI = pd.read_excel('base.xlsx', 'cpi')
dfCPI
```

```python
dfCPI.columns = dfCPI.columns.map(format_columns)
dfCPI.tail()
```

```python
def month(date):
  month = date.month
  if (month < 1 or month > 12):
    return date
  else:
    return month
```

```python
def year(date):
  year = date.year

  return year
```

```python
dfCPI['month'] = dfCPI['date'].apply(month)
dfCPI['year'] = dfCPI['date'].apply(year)
dfCPI
```

```python
dfCPI = dfCPI[['cpiaucsl','varicao12m', 'month','year']]

dfCPI.to_excel('cpi_tratado.xlsx')
```
