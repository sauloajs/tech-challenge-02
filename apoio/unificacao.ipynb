```python
bmp = pd.read_excel('bmp.xlsx')
bmp = bmp[['discriminacao', 'mes', 'inv_carteira', 'inv_direto']]
bmp
```

```python
def inv_carteira(date):
  year = date.year
  month = date.month

  dt = bmp.loc[bmp['discriminacao'] == year]
  dt = dt.loc[dt['mes'] == month]
  values = dt.values

  if (values.any()):
    return float(values[0][2])
  else:
    return None
```

```python
def inv_direto(date):
  year = date.year
  month = date.month

  dt = bmp.loc[bmp['discriminacao'] == year]
  dt = dt.loc[dt['mes'] == month]
  values = dt.values

  if (values.any()):
    return float(values[0][3])
  else:
    return None
```

```python
pib = pd.read_excel('pib.xlsx')
```

```python
pib
```

```python
def get_pib(date, position):
  year = date.year
  month = date.month
  quarter = 0
  matches = pib.loc[pib['ano_pib'] == year]

  if (month <= 3):
    quarter = 3
  elif (month <= 6):
    quarter = 6
  elif (month <= 9):
    quarter = 9
  else:
    quarter = 12

  dt = matches.loc[pib['trimestre_pib'] == quarter]

  values = dt.values

  if (values.any()):
    return float(values[0][position])
  else:
    return None
```
